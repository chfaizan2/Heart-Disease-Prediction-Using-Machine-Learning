# Feature Engineering

```python
# Scaling numerical columns
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,5))
cols = ['age','resting_blood_pressure','cholestoral','Max_heart_rate','oldpeak']
df[cols] = scaler.fit_transform(df[cols])
```

```python
# Performing label encoding
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
cols = df.select_dtypes(include='object').columns
for col in cols:
  df[col] = encoder.fit_transform(df[col])
```

```python
X = df.drop('target', axis=1)
y = df['target']
```

```python
df['target'].value_counts().plot(kind='bar')
plt.title('Target distribution before balancing')
```

```python
# Applying SMOTE to handle imbalanced data
from imblearn.over_sampling import SMOTE
smote = SMOTE()
X_new, y_new = smote.fit_resample(X,y)
```

```python
y_new.value_counts().plot(kind='bar')
plt.title('Target distribution after balancing')
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_new, y_new, test_size=0.2)
```

# Modeling

```python
# Importing Models
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
# Storing functions of models in variable
models = [
    LogisticRegression(),
    DecisionTreeClassifier()
]
```

```python
# Importing Evaluation matrces
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, roc_auc_score
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
# Evaluation function
def evaluate_model(mdl, X_train, y_train, X_test, y_test):

    # Fit model
    mdl.fit(X_train, y_train)

    # Predictions
    y_pred = mdl.predict(X_test)

    # Scores
    train_score = mdl.score(X_train, y_train)
    test_score = mdl.score(X_test, y_test)

    acc = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    name = type(mdl).__name__

    print(f"\n Model: {name}")
    print("Train Accuracy:", train_score)
    print("Test Accuracy :", test_score)
    print("Accuracy      :", acc)
    print("Precision     :", precision)
    print("Recall        :", recall)
    print("F1 Score      :", f1)
    print("\nConfusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))

    # Confusion Matrix Plot
    disp = ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(y_test, y_pred))
    disp.plot(cmap='Blues')
    plt.title(f'Confusion Matrix - {name}')
    plt.show()

    # ROC Curve
    if hasattr(mdl, "predict_proba"):
        y_prob = mdl.predict_proba(X_test)[:, 1]
        roc_auc = roc_auc_score(y_test, y_prob)
        fpr, tpr, _ = roc_curve(y_test, y_prob)

        plt.figure(figsize=(6,5))
        plt.plot(fpr, tpr, label=f'AUC = {roc_auc:.2f}')
        plt.plot([0,1], [0,1], linestyle='--')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(f'ROC Curve - {name}')
        plt.legend()
        plt.show()
# Iterating over a list of models
for mdl in models:
    # Fitting the model to the training data
    mdl.fit(X_train, y_train)

    # Calling the evaluation function
    evaluate_model(mdl, X_train, y_train, X_test, y_test)
```

```python
# Logistic Regression Feature Importance
lr_model = models[0]

feature_importance_lr = pd.Series(
    lr_model.coef_[0],
    index=X_train.columns
).sort_values(key=abs, ascending=False)

feature_importance_lr.plot(kind='bar', figsize=(10,5))
plt.title("Feature Importance - Logistic Regression")
plt.xlabel("Features")
plt.ylabel("Coefficient Value")
plt.show()
```

```python
# Decision Tree Feature Importance
dt_model = models[1]

feature_importance_dt = pd.Series(
    dt_model.feature_importances_,
    index=X_train.columns
).sort_values(ascending=False)

feature_importance_dt.plot(kind='bar', figsize=(10,5))
plt.title("Feature Importance - Decision Tree")
plt.xlabel("Features")
plt.ylabel("Importance Score")
plt.show()
```
